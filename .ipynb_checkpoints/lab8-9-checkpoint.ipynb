```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

**TASK1**

```python
lm1 = LinearRegression()
```

```python
X = np.array(df['engine-size'])
Y = np.array(df['price'])
```

```python
lm1.fit(X.reshape(-1,1), Y)
```

```python
lm1.coef_
```

```python
lm1.intercept_
```

**TASK4**

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Z = df[['normalized-losses', 'highway-mpg']]
```

```python
lm2.fit(Z, df['price'])
```

```python
Y_predict_multifit = lm2.predict(Z)
```

```python
r_squared_multifit = r2_score(df['price'], Y_predict_multifit)
mse_multifit = mean_squared_error(df['price'], Y_predict_multifit)
```

```python
print(f"R-squared ={r_squared_multifit}")
print(f"Mean Squared Error= {mse_multifit}")
print(f'The mean square error of price and predicted value using multifit is: {mse_multifit}')
```
